```python
from matplotlib import pyplot as plt
```

```python


car = ['seltos','alto','maruti','tata','mahendra']
price = [16,6,15,19,20]

plt.plot(car,price,marker='o')

plt.xlabel('car name')
plt.ylabel('car prices')

plt.title('automotive indestries')

plt.show()
```

```python
student = ['ajay','linga','chota don','vijuhatela']
age = [30,28,35,40]

plt.pie(age,labels =student)

plt.title('gnag_student')

plt.show()
```

```python

car = ['seltos','alto','maruti','tata','mahendra']
price = [16,6,15,19,20]

plt.bar(car,price)

plt.xlabel('car name')
plt.ylabel('car prices')

plt.title('automotive indestries')

plt.show()
```
